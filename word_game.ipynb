```python
"""
This is a word guessing game i created with the help of chatGPT. It allows you to paste in a URL, the code then creates a list
of words that appear on that page. The game is to guess the word, entering a letter at a time. Some URL's don't work, i think
this is API related. Wikipedia pages are a good start. 
"""

# import modules needed. "requests" and "beautifulsoup" may need to be installed, use: "pip install requests beautifulsoup4"
import random
import requests
from bs4 import BeautifulSoup

#initialise wordlist
wordlist = []

#function to get words from URL
#i dont fully understand this code right now, basically used as is from CGPT
def get_words_from_web(url):
    response = requests.get(url)
    if response.status_code == 200:
        soup = BeautifulSoup(response.content, 'html.parser')
        # Extract words from the webpage's text
        words = [word.lower() for word in soup.get_text().split() if word.isalpha()]
        return words
    else:
        print("Failed to fetch data from the URL.")
        return []

def game():
    user_url = input("Enter the URL of the webpage containing words: ")
    words_from_web = get_words_from_web(user_url)

    if words_from_web:
        wordlist.extend(words_from_web)
        print("Words added to wordlist.")
        random_index = random.randint(0, len(wordlist) - 1)
    word_to_guess = wordlist[random_index]
    length = len(word_to_guess)
    guessed_word = ["_" for _ in word_to_guess]  # A list to store the correctly guessed letters

    attempts = 0
    max_attempts = len(wordlist[random_index])*2  
    #Sets the maximum number of attempts (i adapted this to make the number
    #of attempts equal to twice the word length )

    while "_" in guessed_word and attempts < max_attempts:
        guess = input("Enter your guess for a letter: ").lower() #lower means user can use any case for guess

        if len(guess) != 1 or not guess.isalpha(): # check if all the characters in a string are alphabetic characters
            print("Invalid input. Please guess a single letter.")
            continue

        if guess in guessed_word:
            print("You already guessed this letter.")
            continue

        if guess in word_to_guess:
            for i, letter in enumerate(word_to_guess):
                if letter == guess:
                    guessed_word[i] = guess
            print("Correct guess!")
        else:
            print("Incorrect guess!")

        print(" ".join(guessed_word))
        attempts += 1

    if "_" not in guessed_word:
        print("Congratulations! You guessed the word correctly!")
    else:
        print("Sorry, you ran out of attempts. The word was:", word_to_guess)

#run the game
game()
```
